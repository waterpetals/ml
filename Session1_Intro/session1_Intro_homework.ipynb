<h2> Question 7 </h2>

```python
island=df[df['ocean_proximity']=='ISLAND']
island.head()
```

```python
X=island[['housing_median_age', 'total_rooms', 'total_bedrooms']]
X
```

```python
XT = X.T
XT
```

```python
X.shape
```

```python
XT.shape
```

```python
XTX = np.dot(XT, X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)

XTX_inv, XTX_inv.sum()
```

```python
y = [950, 1300, 800, 1000, 1300]
w= np.dot(np.dot(XTX_inv, XT), y)
w
```
